```python
import matplotlib.pyplot as plt
import pandas as pd
from   datetime import datetime
import numpy as np
import sys, getopt

# Keyboard shortcuts
# https://www.cheatography.com/weidadeyue/cheat-sheets/jupyter-notebook/
# https://pandas.pydata.org/pandas-docs/stable/10min.html

def drawTimeSeries ( ppp, data, pSeries, yLabel, plotTitle, colKeys):

   # Keep a record of the initial column keys
   dfEndpoint = data.shape[0] - 1
   ppp.plot( data.values[1:dfEndpoint,0], data.values[1:dfEndpoint,pSeries] )
   #ppp.gcf().autofmt_xdate()
   ppp.ylabel(yLabel)
   perlRocks = "%s %d" % (plotTitle, dfEndpoint)
   ppp.title(perlRocks)
   #print (data)
   ppp.show()
```
